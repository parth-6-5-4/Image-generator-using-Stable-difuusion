```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "stabilityai/stable-diffusion-2-1"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
image_model= "dreamlike-art/dreamlike-diffusion-1.0"
```

```python
prompt="""dreamlike a deer in the forest besides a lake with mountains in the backdrop of a sunset with orange and purple hues"""
```

```python
import matplotlib.pyplot as plt
```

```python
image= pipe(prompt).images[0]
image.save("image.png")
plt.imshow(image)
plt.axis("off")
plt.show
```

```python
prompt="""dreamlike a deer in the forest  with mountains in the backdrop of a sunset with orange and purple hues in polygon art format"""
image= pipe(prompt).images[0]
image.save("image.png")
plt.imshow(image)
plt.axis("off")
plt.show
```

##Parameters in stable diffusion pipeline
1. num_inference_steps
2. negative prompting
3. height
4. width
5. number of images per prompt(num_images_per_prompt)

```python
prompt_1= """an astronaut petting a deer in a forest where it is snowing after crash landing with the crashed spacecraft in the backdrop a bit far , blue and red hues"""

images= pipe(prompt_1, num_inference_steps=100, num_images_per_prompt=2).images
plt.imshow(images[0],images[1])
plt.axis("off")
plt.show
```
